## 我以以下程序断言:除了个别列缺少值以外,两个数据集并无不同
### 结论:不采用新的数据集,使用旧的数据集

```python
KDDTest = pd.read_csv('KDDTest.csv',dtype='str')
KDDTrain = pd.read_csv('KDDTrain.csv',dtype='str')
KDDTest_ = pd.read_csv('KDDTest+.txt',dtype='str')
KDDTrain_ = pd.read_csv('KDDTrain+.txt',dtype='str')
```

```python
#对数据集进行逐行对比(数据集中的数据类型为str)
def datacompare(data1,data2):
    #将数据集转换为list
    data1 = data1.values.tolist()
    data2 = data2.values.tolist()
    #对数据集进行逐行对比
    for i in range(len(data1)):
        if data1[i] != data2[i]:
            print(i)
            print(data1[i])
            print(data2[i])
            print('------------------')
    print('数据集对比完毕！')
```

```python
datacompare(KDDTest,KDDTest_)
```
